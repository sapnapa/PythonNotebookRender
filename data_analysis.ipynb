```python
xl = pd.io.excel.ExcelFile("C:\\Users\\namitk\\Documents\\CodeAnalysis.xlsx")
```

```python
sheet1 = xl.parse("Sheet1")
sheet2 = xl.parse("Sheet2")
```

```python
sheet2.head(5)
```

```python
_df_1 = sheet2[['Rating']].groupby("Rating").count()
ax = _df_1.plot(kind="bar", legend=True, color="#778888")
ax.set_title("Rating count", fontsize=20)
ax.tick_params(axis='both', colors='black', labelsize=20)
ax.set_xlabel("Rating", fontsize=20)
ax.set_ylabel("Count", fontsize=20);
```

```python
groups = sheet2.groupby('Rating')
```

```python
colors = pd.tools.plotting._get_standard_colors(len(groups), color_type='random')
```

```python
for name, group in groups:
    ax_1 = group.plot()
    ax_1.set_ylabel("issue counts")
    ax_1.set_title("for " + name + " Rating ", fontsize=20)
    ax_1.tick_params(axis='both', colors='black', labelsize=20)
    ax_1.set_xlabel("issues", fontsize=20)
    ax_1.set_ylabel("Issue counts", fontsize=20)    
plt.show()
```
